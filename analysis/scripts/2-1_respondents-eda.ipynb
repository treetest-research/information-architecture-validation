### Education level

```python
my_bar(respondents, 'education', 'What is the highest education level you have completed?', None, None, 
    ['None completed', 'Secondary education', 'High school diploma', 'Technical/community college', 'Undergraduate degree', 'Graduate degree', 'Doctorate degree']
)
```

```python
my_bars(respondents, 'education', figSize=(30, 15), 
    order=['None completed', 'Secondary education', 'High school diploma', 'Technical/community college', 'Undergraduate degree', 'Graduate degree', 'Doctorate degree']
)
```

```python
# income
my_bar(respondents, 'income', 'What is your personal income per year, after tax?', None, None, 
       ['No income', '0 - 9,999', '10,000 - 19,999', '20,000 - 29,999', '30,000 - 39,999', '40,000 - 49,999', '50,000 - 74,999', '75,000 - 100,000', 'More than 100,000'])
```

```python
# web browsing
my_bar(respondents, 'web', 'How often do you browse the web?', None, None, 
    ['Never', 'Very sporadically (few times a month)', 'From time to time (few times a week)', 'Few times a day', 'For the major part of the day']
)
```

```python
respondents[['variant', 'web']].groupby(['variant', 'web']).size().unstack().fillna(0)
```

```python
my_bars(respondents, 'web', figSize=(30, 15),
    order=['Never', 'Very sporadically (few times a month)', 'From time to time (few times a week)', 'Few times a day', 'For the major part of the day']
)
```

```python
# magazine browsing
my_bar(respondents, 'magazine', 'How often do you visit online magazines or news websites?', None, None, 
    ['Never', 'Very sporadically (monthly)', 'From time to time (weekly)', 'Daily']
)
```

```python
respondents[['variant', 'magazine']].groupby(['variant', 'magazine']).size().unstack().fillna(0)
```

```python
my_bars(respondents, 'magazine', figSize=(35, 15), 
    order=['Never', 'Very sporadically (monthly)', 'From time to time (weekly)', 'Daily']
)
```
