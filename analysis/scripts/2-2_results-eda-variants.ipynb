## Changeminds

```python
results['changemind'] = results['changemindCount'] = None

for index, row in results.iterrows():
    last_in_path = list(map(lambda x: x[-1] if len(x)>0 else 'H', json.loads(row.path)))
    s = sum(list(map(lambda x: 1 if x in LEAF_NODES else 0, last_in_path))[0:-1])
    results.loc[index, 'changemind'] = s > 0
    results.loc[index, 'changemindCount'] = s
```
